```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv("assn3.csv",sep='\t')
```

```python
df.head()
```

```python
records = []
for i in range(0,21):
    records.append([str(df.values[i,j]) for j in range(0,6)])
```

```python
new_records = []
for i in range(0,21):
    temp = []
    for j in range(0,len(records[i])):
        if(records[i][j]!='nan'):
            temp.append(records[i][j])
    new_records.append(temp)
```

```python
print(records)
print(len(records))
```

```python
association_rules = apriori(records,min_support = 0.095238095238,min_confidence=0.6,min_lift=1.1,min_length=2)
association_results = list(association_rules)
```

```python
print(association_results)
```

```python
print(len(association_results))
```
