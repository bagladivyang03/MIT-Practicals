```python
data.drop([13,14,15,16],axis=1,inplace=True)
```

```python
records1 = []
for i in range(0,1009):
    records1.append([str(data.values[i,j]) for j in range(0,13)])
new_records1 = []
```

```python
for i in range(0,1009):
    temp = []
    for j in range(0,len(records1[i])):
        if(records1[i][j]!='nan'):
            temp.append(records1[i][j])
    new_records1.append(temp)
```

###### Implementing the FP Growth Algorithm

##### Calculating the time of execution

```python
start = time.time()
patterns = pyfpgrowth.find_frequent_patterns(new_records1,4.5)
rules = pyfpgrowth.generate_association_rules(patterns,0.6)
print(time.time()-start)
```

Here,

find_frequent_patterns is used to find patterns in baskets that occur over the support threshold.

generate_association_rules is used to find patterns that are associated with another with a certain minimum probability.
